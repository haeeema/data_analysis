### 영등포구 소재 지하철역

```python
import folium
import pandas as pd
```

```python
df = pd.read_csv("data/영등포소재 지하철역2.csv")
df
```

```python
map = folium.Map(location=[df['위도'].mean(), df['경도'].mean()], zoom_start=14)    # Center position
for i in df.index:
    folium.Marker(
        location=[df['위도'][i], df['경도'][i]],
        tooltip=df['역명'][i],
        popup=folium.Popup(df['주소'][i], max_width=300),
    ).add_to(map)
title_html = '<h3 align="center" style="font-size:20px"><b>영등포소재 주요 지하철역</b></h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```

```python
df['color'] = ['red', 'blue', 'green', 'violet', 'orange']
df
```

```python
map = folium.Map(location=[df['위도'].mean(), df['경도'].mean()], zoom_start=14)    # Center position
for i in df.index:
    folium.Circle(
        location=[df['위도'][i], df['경도'][i]],
        tooltip=df['역명'][i],
        popup=folium.Popup(df['주소'][i], max_width=300),
        radius=100, color=df['color'][i], fill=True,
    ).add_to(map)
    folium.Marker(
        location=[df['위도'][i], df['경도'][i]],
        icon=folium.DivIcon(icon_size=(80,20), icon_anchor=(20,0),
                            html=f'<span style="font-size: 10pt">{df.역명[i]}</span>')
    ).add_to(map)

title_html = '<h3 align="center" style="font-size:20px"><b>영등포소재 주요 지하철역</b></h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```
