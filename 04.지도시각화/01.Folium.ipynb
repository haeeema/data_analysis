### Folium 지도 시각화
- 설치
    - conda activate base
    - pip install folium

```python
import folium
folium.__version__
```

- map 객체 만들기

```python
# Open Street map
map = folium.Map(location=[37.51528, 126.9075611], zoom_start=15)       # 영등포역
map
```

- tileset 선택

```python
# default tileset: Open Street Map
# Stamen Toner: 단계구분도(Choropleth)를 그릴 때 사용
map = folium.Map(location=[37.51528, 126.9075611], zoom_start=15, tiles='Stamen Toner')       # 영등포역
map
```

- Marker 추가

```python
map = folium.Map(location=[37.51528, 126.9075611], zoom_start=13)
folium.Marker(
    location=[37.51528, 126.9075611],    # WGS84 좌표계: 위도, 경도
    tooltip='영등포역',
    popup=folium.Popup('서울특별시 영등포구 경인로 846 (영등포본동)', max_width=300),
).add_to(map)
folium.Marker(
    location=[37.52272, 126.905131],
    tooltip='영등포시장역',
    popup=folium.Popup('서울특별시 영등포구 양산로 지하200 (영등포동5가 62-1)', max_width=300),
).add_to(map)
folium.Marker(
    location=[37.5320028, 126.9140722],
    tooltip='국회의사당',
    popup=folium.Popup('서울특별시 영등포구 의사당대로 1', max_width=300),
).add_to(map)
folium.Marker(
    location=[37.526344, 126.896256],
    tooltip='영등포구청',
    popup=folium.Popup('서울특별시 영등포구 당산동3가 당산로 123', max_width=300),
).add_to(map)
map
```

- 다양한 Marker

```python
map = folium.Map(location=[37.52028, 126.9075611], zoom_start=14)
folium.Marker(
    location=[37.51528, 126.9075611],
    tooltip='영등포역',
    popup=folium.Popup('서울특별시 영등포구 경인로 846 (영등포본동)', max_width=300),
).add_to(map)
folium.Marker(
    location=[37.52272, 126.905131],
    tooltip='영등포시장역',
    popup=folium.Popup('서울특별시 영등포구 양산로 지하200 (영등포동5가 62-1)', max_width=300),
    icon=folium.Icon(color='green')
).add_to(map)
folium.Circle(
    location=[37.5320028, 126.9140722],
    tooltip='국회의사당',
    radius=100, color='crimson', fill=True, fill_color='crimson',
    popup=folium.Popup('서울특별시 영등포구 의사당대로 1', max_width=300),
).add_to(map)
folium.CircleMarker(
    location=[37.526344, 126.896256],
    tooltip='영등포구청',
    radius=13, color='#3186cc', fill=True,
    popup=folium.Popup('서울특별시 영등포구 당산동3가 당산로 123', max_width=300),
).add_to(map)
title_html = '<h3 align="center" style="font-size:20px"><b>학원근처 유명 지점</b></h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```
