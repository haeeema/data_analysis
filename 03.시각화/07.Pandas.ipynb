### Pandas의 시각화 기능

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

- Line graph

```python
np.random.seed(0)
df = pd.DataFrame(np.random.randn(100, 3), columns=['A', 'B', 'C'],
                  index=pd.date_range('2023-09-01', periods=100))
df.tail()
```

```python
df = df.cumsum()
df.tail()
```

```python
df.plot(figsize=(5, 3))         # default: line plot
plt.title('Pandas의 plot 메소드 사용 예')
plt.xlabel('시간'), plt.ylabel('Data')
plt.show()
```

- Bar chart

```python
iris = sns.load_dataset('iris')

iris.sepal_length[:30].plot(kind='bar', rot=45, figsize=(6, 3))
plt.title('tirty data of iris sepal_length');
```

```python
iris.sepal_length[:30].plot(kind='barh')
plt.title('tirty data of iris sepal_length');
```

```python
iris.sepal_length[:30].sort_values(ascending=False).plot(kind='barh')
plt.title('tirty data of iris sepal_length');
```

```python
df2 = iris.groupby(iris.species).mean()
df2
```

```python
df2.plot.bar(figsize=(6, 3), rot=0)
plt.title('mean data of iris species');
```

```python
df2.columns.name = 'Feature'
df2.plot.bar(figsize=(6, 3), rot=0, colormap='viridis')
plt.ylim(0, 10)
plt.title('mean data of iris species');
```

- Pie chart

```python
titanic = sns.load_dataset('titanic')

s = titanic['class'].value_counts()
s
```

```python
s.sort_index().plot.pie(autopct='%.1f%%')
plt.title('선실별 승객수');
```

- Histogram

```python
titanic.age.plot.hist(bins=30, figsize=(6, 3))
plt.title('Histogram of titanic age');
```

- Box plot

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.boxplot(iris[iris.species == 'setosa'].sepal_length)
plt.subplot(1, 2, 2)
plt.boxplot(iris[iris.species == 'versicolor'].sepal_length)
plt.show()
```

```python
iris[iris.species == 'setosa'].plot.box(figsize=(5, 3))
plt.title('setosa 품종의 각 feature별 box plot');
```

```python
iris.boxplot(by='species', figsize=(10, 6))
plt.suptitle('각 feature별 box plot');
```
