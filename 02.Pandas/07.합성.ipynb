### 2.7 데이터프레임 합성(merge)

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
# Inner join - 교집합
pd.merge(df1, df2)
```

```python
# left outer join - 왼쪽 데이터프레임을 모두 포함
pd.merge(df1, df2, how='left')
```

```python
# Right outer join
pd.merge(df1, df2, how='right')
```

```python
# Full outer join
pd.merge(df1, df2, how='outer')
```

- join() method

```python
# index를 기준으로 join
df1.set_index('고객번호', inplace=True)
df2.set_index('고객번호', inplace=True)
```

```python
df1.join(df2)   # Left outer join
```

```python
# Inner join
pd.merge(df1, df2, left_index=True, right_index=True)
```

- 연습문제 4.6.1

```python
df3 = pd.DataFrame(np.random.randint(100, 110, size=(5, 5)), columns=list('ABCDE'))
df3
```

```python
df4 = pd.DataFrame(np.random.randint(100, 110, size=(5, 5)), columns=list('ABXYZ'))
df4
```

```python
pd.merge(df3, df4, on='A')
```

```python
pd.merge(df3, df4, on='B')
```

```python
# A열과 B열을 기준으로 합치기
pd.merge(df3, df4, on=['A', 'B'])
```

```python
# C열 == X열 기준
pd.merge(df3, df4, left_on='C', right_on='X')
```

- concat

```python
df5 = pd.DataFrame(
    np.arange(6).reshape(3, 2),
    index=['a', 'b', 'c'],
    columns=['데이터1', '데이터2'])
df5
```

```python
df6 = pd.DataFrame(
    5 + np.arange(4).reshape(2, 2),
    index=['a', 'c'],
    columns=['데이터3', '데이터4'])
df6
```

```python
pd.concat([df5, df5])   # Vertical stack
```

```python
pd.concat([df5, df6], axis=1)   # Horizontal stack
```
